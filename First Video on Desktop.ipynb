```python
# This script finds the first video file on the current user's Desktop
# and opens it using the default associated application.

# 1. Get the path to the Desktop folder
$desktopPath = [System.Environment]::GetFolderPath('Desktop')

# 2. Define common video file extensions to look for
$videoExtensions = @('.mp4', '.mkv', '.avi', '.mov', '.wmv', '.flv', '.webm', '.m4v', '.mpg', '.mpeg')

# 3. Search for files on the Desktop
Write-Host "Searching for videos on Desktop..." -ForegroundColor Cyan

$firstVideo = Get-ChildItem -Path $desktopPath -File |
    Where-Object { $videoExtensions -contains $_.Extension.ToLower() } |
    Select-Object -First 1

# 4. Check if a video was found and play it
if ($firstVideo) {
    Write-Host "Found video: $($firstVideo.Name)" -ForegroundColor Green
    Write-Host "Playing now..." -ForegroundColor Green
    Invoke-Item -Path $firstVideo.FullName
} else {
    Write-Host "No video files found on the Desktop." -ForegroundColor Yellow
}
```
