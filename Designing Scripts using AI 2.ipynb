This is a great choice. **Profile C ("The Assistant")** strikes the perfect balance between automation and control. It keeps your Windows 11 machine clean but ensures you never accidentally delete a file you meant to recover.

Here is the detailed design specification for your **Windows 11 Recycle Bin Assistant**.

---

### 1. The Core Logic

We will use **PowerShell**, which is native to Windows 11. It allows us to easily calculate storage size, display Windows-style popups, and interact with the system shell.

**The Workflow:**

1. **Scan:** Upon double-clicking the script, it silently scans the Recycle Bin.
2. **Calculate:** It sums up the **Total File Count** and **Total Size** (converted to MB or GB for readability).
3. **Report & Ask:** It generates a popup window (Message Box) centered on your screen.
* *Message:* "Your Recycle Bin contains **15 files** totaling **1.2 GB**. Do you want to empty it?"
* *Buttons:* `Yes` / `No`


4. **Decision:**
* **If No:** The script closes immediately. Nothing happens.
* **If Yes:** The script executes the `Clear-RecycleBin` command.


5. **Feedback:** (Optional) A small "toast" notification or sound plays to confirm the bin is empty.

### 2. Technical Components

#### A. Gathering Statistics

To provide the "Assistant" experience, the script needs to know *what* is in the bin before deleting it.

* We will use the PowerShell cmdlet `Get-ChildItem` pointed at the Recycle Bin directory.
* We will pipe this into `Measure-Object` to sum the `Length` (file size).
* *Constraint:* We must handle permission errors gracefully (e.g., if the script cannot read a specific system file in the bin, it shouldn't crash).

#### B. The User Interface (GUI)

Instead of a scary black command prompt window, we will use the **.NET Framework** classes built into Windows to show a clean, graphical message box.

* **Class:** `[System.Windows.Forms.MessageBox]`
* **Icon:** We will set the icon to `Question` or `Warning` to grab your attention.
* **Window Style:** It will look like a standard Windows 11 system prompt.

#### C. The Cleaning Engine

We will use the modern Windows command:

* `Clear-RecycleBin`
* **Parameters:** We will use `-Force` (to bypass the *second* Windows confirmation, since our custom popup already asked you) and `-ErrorAction SilentlyContinue` (to ignore locked files that cannot be deleted).

### 3. User Experience (The "Trigger")

You don't want to open a code editor to run this. The design includes creating a **Desktop Shortcut**.

* **The Icon:** We can set the shortcut icon to look like a trash can or a broom (using Windows standard icons).
* **The Execution:** You will simply double-click this icon.
* **Visibility:** The script will be designed to run "hidden" or minimized, so you only see the Popup Box, not the underlying code window.

---

### 4. Safety Features specific to "The Assistant"

1. **The "Empty" Check:** If the script runs and the bin is *already* empty, it should not bother you with a "Delete 0 files?" question. It should simply tell you "Bin is already clean" and close.
2. **Focus:** The popup box must force itself to the foreground (be the "TopMost" window) so you don't miss it behind your web browser.

---

### Summary of the Design

* **Platform:** Windows 11 (PowerShell 5.1 or Core 7+)
* **Input:** Double-click on Desktop Icon.
* **Interaction:** "Bin contains X files (Y MB). Proceed?"
* **Safety:** Requires explicit "Yes" click; Checks permissions.
